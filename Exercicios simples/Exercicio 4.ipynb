  <dt>np.random.rand(10,20):</dt>
  <dd> cria um array 2d com 20 elementos por posição e 10 posições ao todo. Os elementos são números entre 0 e 1 apenas. Quando multiplicados por 8, os números passam a ter a casa dos inteiros e não somente de float.</dd>
  <dt>astype():</dt>
  <dd>converte os dados para o tipo que você configurar, nesse caso números inteiros.</dd>
  <dt>plt.imshow(array,'gray'):</dt>
  <dd>Esse comando transforma a matriz "array" com os valores inteiros em um gráfico-imagem com escala de cinza (gray)</dd>
  <dt>plt.show():</dt>
  <dd>Esse comando mostra o grafico-imagem que foi criado</dd>
